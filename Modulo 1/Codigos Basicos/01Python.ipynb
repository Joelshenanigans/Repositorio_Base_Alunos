# O que são variavéis

Variavéis são espaços na memória do computador onde podemos guardar informações (como textos, números, resultados de cálculo, etc) para usar depois  

Variável com texto (string)

```python
nome = "joel"
print(nome)
```

```python
nome = "brando"
```

```python
print(nome)
```

Variavél com número inteiro (int)

```python
idade = 16
print(idade)
```

```python
temperatura = -3 # esta variavel está armazenando um numero negativo
print(temperatura) # print () é uma função de saída
```

Variável com número decimal (float)

```python
altura = 1.79 #Use "." para números decimais em python
print(altura)
```

```python
pi = 1.79
print(pi)
```

Variável com valor booleano(bool)

```python
estudante = True
print(estudante)
```

# Nome de Variavéis

Pode:

*   Começar com letra Exemplos( n, numero, altura)
*   Conter número, mas não pode começar com número Exemplo(n1, numero1, etc)

Não pode:


*   começar com número
*   usar espaço Exemplo(nome_completo, nome_completo,nomeCompleto)
*   usar acentuação ou caracteres especiais(%, &, #, @)
*   usar palavras reservadas







# O que é snake case?

snake_case é um estilo é um estilo de escrita usado para nome de variáveis, funções, e nomes em geral no código. Neste estlo, as palavras são todas em minúsculas e separadas por um undeline(_)

#Porque snake_case


* Deixa o código fácil de ler
* É o padra oficial recomendado pela python
* Segue o PEF8, que é o guia de estilo oficial do python

# Exemplos que não usam no snake_case
