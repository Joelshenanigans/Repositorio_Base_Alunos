# Função Print()
a função print() é uma das mais importantes e usadas na linguagem Python. Sua função é, basicamente, exibir mensagens na tela ou enviá-las para outro dispositivo, como imprimir dentro de um arquivo de texto

No Python 3, print() é uma função nterna, de modo que não é necessário importar nenhuma biblioteca para poder utiliza-lá. Basta chama-lá e passar os argumentos necessários

```python
print("Meu primeiro print") #todo texto deve estar entre aspas simples ou duplas
```

```python
print("Meu segundo print")
```

```python
print("D'água") # quando uma palavra contém aspas, as aspas iniciais devem ser diferentes
```

```python
print(20 + 15) # O Python realiza a operação que está dentro do parenteses
```

```python
print("20+15") # a operação entre aspas sera apenas impressa não executada, pois o python entende isso como texto
```

```python
print("20 + 15 = ", 20 + 15)
```

```python
n1 = "1"
n2 = "3"
print(n1 + n2) # neste exemplo o python une as 2 strings
```

```python
nome = "The"
sobrenome = "World"
nome_completo = nome + " "+ sobrenome
print(nome_completo)
```
